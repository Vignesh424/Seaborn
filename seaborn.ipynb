```python
#SEABORN Line Chart
import seaborn as sns
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
#SWARM PLOTS
import seaborn as sns
sns.set_theme(style="whitegrid", palette="muted")

# Load the penguins dataset
df = sns.load_dataset("penguins")

# Draw a categorical scatterplot to show each observation
ax = sns.swarmplot(data=df, x="body_mass_g", y="sex", hue="species")
ax.set(ylabel="")
```

```python
#IMPLOT 
import seaborn as sns
sns.set_theme(style="ticks")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")

# Show the results of a linear regression within each dataset
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

```python
#HISTOGRAM
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'],kde = False)
plt.show()
```

```python
#DISTPLOT with Histogram
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'],hist=True)
plt.show()
```

```python
#DISTPLOT without Histogram
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'],hist=False)
plt.show()
```

```python
#JOINTPLOT
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.jointplot(x = 'petal_length',y = 'petal_width',data = df)
plt.show()
```

```python
#STRIP PLOT WITH JITTER
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.stripplot(x = 'petal_length',y = 'petal_width',data = df, jitter=True)
plt.show()
```

```python
#BAR PLOT
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('penguins')
ax = sns.barplot(data=df, x="body_mass_g", y="sex", hue="species")
ax.set(ylabel="")
```

```python
#BOX PLOTS
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('penguins')
ax = sns.boxplot(data=df, x="body_mass_g", y="sex", hue="species")
ax.set(ylabel="")
```

```python
#VIOLIN PLOTS
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset("penguins")
ax = sns.violinplot(data=df, x="body_mass_g", y="sex", hue="species")
ax.set(ylabel="")
```

```python
#POINT PLOTS
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('titanic')
sb.pointplot(x = "sex", y = "survived", hue = "class", data = df)
plt.show()
```
